```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('AirQualityUCI.xlsx')
```

```python
# Displaying the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Displaying basic information about the dataset
print("\nBasic Information:")
print(df.info())
```

```python
# Displaying summary statistics of the dataset
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Displaying datatypes
print("\nData Types:")
print(df.dtypes)
```

```python
# Displaying unique values count
print("\nUnique Values Count:")
print(df.nunique())
```

```python
# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Visualizing missing values using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Distribution of Temperature (T)
plt.figure(figsize=(10, 6))
sns.histplot(df['T'], kde=True, bins=30)
plt.title('Distribution of Temperature (T)')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot for selected variables
selected_columns = ['T', 'RH', 'AH', 'CO(GT)', 'NO2(GT)']
sns.pairplot(df[selected_columns])
plt.show()
```

```python
# Plotting histograms for all numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns].hist(figsize=(12, 10), bins=30)
plt.suptitle('Histograms of Numeric Columns')
plt.show()
```
